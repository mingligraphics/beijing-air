```python
df_2015_pm2 = pd.read_csv('2015_pm2.csv', parse_dates = ['date'])
```

```python
df_2015_pm2.head()
```

```python
def f(df):
    df = df.copy()
    df['year'] = pd.DatetimeIndex(df['date']).year
    df['month'] = pd.DatetimeIndex(df['date']).month
    df['day'] = pd.DatetimeIndex(df['date']).day
    return df
```

```python
df_2015_pm2 = f(df_2015_pm2)
```

```python
df_2015_pm2.month.value_counts()
```

```python
del df_2015_pm2['hour']
```

```python
def label_quarter (row):
   if row['month'] == 1 or row['month'] == 2 or row['month'] == 3 :
      return 1
   if row['month'] == 4 or row['month'] == 5 or row['month'] == 6 :
      return 2
   if row['month'] == 7 or row['month'] == 8 or row['month'] == 9 :
      return 3
   if row['month'] == 10 or row['month'] == 11 or row['month'] == 12 :
      return 4
```

```python
df_2015_pm2['quarter'] = df_2015_pm2.apply (lambda row: label_quarter(row), axis=1)
```
