```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_csv('pm2_5.csv')
```

```python
df.plot(x='Quarter',y='level')
```

```python
df2 = pd.read_csv('pm2_5_month.csv')
```

```python
ax = df2.plot(x='month', y='level',marker='o', markerfacecolor="white", fillstyle='none',figsize=(24,8))
ax.set_ylim(0,180)

df2.plot(ax=ax, x='month', y='WHO', markerfacecolor="white")
ax.grid(linestyle='--', color='lightgrey')
ax.tick_params(which='both', # Options for both major and minor ticks
                top='off', # turn off top ticks
                left='off', # turn off left ticks
                right='off',  # turn off right ticks
                bottom='off')
plt.savefig("linechart1.pdf")
```
